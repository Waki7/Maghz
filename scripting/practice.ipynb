```python
from torch.utils.data import DataLoader
from typing import *


class ExperimentTracker:
    def __init__(self):
        self.writer = SummaryWriter()

    def log_scalar(self, name: str, value: float, step: Optional[int] = None):
        self.writer.add_scalar(name, value, step)



class Data:
    def __init__(self, data: TensorT['B,S']):
        self.train = torch.rand(100, 3)
        self.test = torch.rand(100, 3)


    def collate_fn(self, batch: List[Tuple[TensorT['S'], TensorT['S']]]) -> Tuple[TensorT['B,S'], TensorT['B,S']]:
        return torch.stack([x[0] for x in batch]), torch.stack([x[1] for x in batch])


data_loader = DataLoader(Data(), batch_size=32, collate_fn=Data.collate_fn)
```
